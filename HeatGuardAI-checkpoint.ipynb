```python
# ========================================================================================
# STEP 1: Columns in Heat Stroke Dataset
# ========================================================================================
print("Heat Stroke Dataset Columns:")
print(df_heat.columns.tolist())

# ========================================================================================
# STEP 2: Identify categorical and numerical columns
# ========================================================================================
categorical_cols_heat = [
    'Cardiovascular disease history', 'Dehydration', 'Sickle Cell Trait (SCT)',
    'Exposure to sun', 'Exertional (1) vs classic (0)', 'Sweating',
    'Skin color (flushed/normal=1, pale=0.5, cyatonic=0)',
    'Strenuous exercise', 'Nationality', 'Sex', 'Hot/dry skin', 'Time of day'
]

target_col_heat = 'Heat stroke'

# numerical = all remaining except target
numerical_cols_heat = [col for col in df_heat.columns 
                       if col not in categorical_cols_heat + [target_col_heat]]

print("\nCategorical Columns (Heat Stroke):", categorical_cols_heat)
print("Numerical Columns (Heat Stroke):", numerical_cols_heat)

# ========================================================================================
# STEP 3: Encode categorical columns
# ========================================================================================
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in categorical_cols_heat:
    df_heat[col] = le.fit_transform(df_heat[col])

# ========================================================================================
# STEP 4: Features and Target
# ========================================================================================
X_heat = df_heat.drop(target_col_heat, axis=1)
y_heat = df_heat[target_col_heat]

# ========================================================================================
# STEP 5: Train-Test Split
# ========================================================================================
from sklearn.model_selection import train_test_split

X_train_heat, X_test_heat, y_train_heat, y_test_heat = train_test_split(
    X_heat, y_heat, test_size=0.2, random_state=42
)

# ========================================================================================
# STEP 6: Scale numerical features
# ========================================================================================
from sklearn.preprocessing import StandardScaler

scaler_heat = StandardScaler()
X_train_heat[numerical_cols_heat] = scaler_heat.fit_transform(X_train_heat[numerical_cols_heat])
X_test_heat[numerical_cols_heat] = scaler_heat.transform(X_test_heat[numerical_cols_heat])

print("\n✅ Heat Stroke dataset preprocessing completed!")

# ========================================================================================
# STEP 7: Display Preprocessed Training Data (like Traffic_Behavior notebook output)
# ========================================================================================

# Convert back to DataFrame for clean display
X_train_heat_display = pd.DataFrame(X_train_heat, columns=X_heat.columns)
X_test_heat_display  = pd.DataFrame(X_test_heat,  columns=X_heat.columns)

print("\n📊 Sample Preprocessed Training Data:")
X_train_heat_display.head(15)
```
